```python
#Import libraries
import numpy as np
import pandas as pd
import math
import random
import seaborn as sns
import matplotlib.pyplot as plt
import os

```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")

print("Path to dataset files:", path)
```

```python
print(os.listdir(path))
```

```python
df = pd.read_csv(os.path.join(path, 'tmdb_5000_movies.csv'))
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[['id', 'title', 'genres', 'budget', 'popularity', 'release_date', 'revenue', 'vote_average']]
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# Check if any rows have 'id' equal to 1
if df[df['id'] == 28].empty:
    print("No rows found with id equal to 1.")
else:
    # If rows with 'id' equal to 1 exist, access the 'genres' column
    print(df.loc[df['id'] == 28, 'genres'].iloc[0])
```

```python
print(df['genres'].iloc[1])
```

```python
import json

df['genres'] = df['genres'].apply(json.loads)
```

```python
df['genre_names'] = df['genres'].apply(lambda x: [d['name'] for d in x])

df_exploded = df.explode('genre_names', ignore_index=True)
```

```python
df_exploded.head()
```

```python
df = df_exploded.drop('genres', axis=1)
```

```python
df['gross'] = df['revenue'] - df['budget']
```

```python
df.head(10)
```

```python
# Normalisasi kolom popularity ke skala 1-10
min_pop = df['popularity'].min()
max_pop = df['popularity'].max()

df['popularity_scaled'] = (1 + ((df['popularity'] - min_pop) / (max_pop - min_pop)) * (100 - 1)).round(1)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df['title_year'] = df['release_date'].dt.year
```

```python
df.head()
```

```python
df.info()
```

```python
movie_metadata = pd.read_csv("movie_metadata.csv")
movie_metadata.info()
```

```python
movie_metadata = movie_metadata[['title_year', 'imdb_score']]
```

```python
# Merge on a common column (e.g., 'id')
merged_df = pd.merge(df, movie_metadata, on='title_year')  # Default is an inner join

# Save the merged data to a new CSV
merged_df.to_csv('merged_movie_data.csv', index=False)

print("Merged data saved to 'merged_movie_data.csv'")
```

```python
df = pd.read_csv('merged_movie_data.csv')
```

```python
df.head()
```

```python
df = df[['title_year', 'gross', 'popularity_scaled', 'imdb_score', 'vote_average', 'genre_names']]
```

```python
df = df.rename(columns={'title_year': 'year'})
df = df.rename(columns={'genre_names': 'genre'})
df = df.rename(columns={'popularity_scaled': 'popularity'})
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['year'] = df['year'].astype(int)
```

```python
# Simpan data yang sudah ada label
df.to_csv('movies_dataset_summary.csv')
```
